```python
import matplotlib.pyplot as plt
import numpy as np

data1=np.random.normal(0,1,100)
data2=np.random.normal(1,2,100)
data3=np.random.normal(2,1.5,100)

#create sub-plots
fig,axs=plt.subplots(1,3,figsize=(15,5))

#PLOT FOR DATA 1
axs[0].boxplot(data1)
axs[0].set_title('Data 1')
axs[0].set_xlabel('SAMPLE')
axs[0].set_ylabel('VALUES')

#PLOT FOR DATA2

axs[1].boxplot(data2)
axs[1].set_title('Data 2')
axs[1].set_xlabel('SAMPLE')
axs[1].set_ylabel('VALUES')

#PLOT FOR DATA 3

axs[2].boxplot(data3)
axs[2].set_title('Data 3')
axs[2].set_xlabel('SAMPLE')
axs[2].set_ylabel('VALUES')

#adjust layout

plt.tight_layout()
plt.show()
```

```python
data_group1=[np.random.normal(0,1,100),np.random.normal(1,2,100),np.random.normal(2,1.5,100)]
data_group2=[np.random.normal(0,1,100),np.random.normal(1,2,100),np.random.normal(2,1.5,100)]

data=data_group1+data_group2

#creating grouped boxplots

plt.boxplot(data,positions=[1,2,3,5,6,7],labels=['G1-D1','G1-D2','G1-D3','G2-D1','G2-D2','G2-D3'])
plt.title('GROUPED BOXPLOTS')

plt.xlabel('GROUPED DATA-SET')

plt.ylabel('VALUE')

plt.show()
```

```python
#give colours
plt.boxplot(data,boxprops=dict(color='BLUE'))
plt.show()
```

```python
plt.boxplot(data,whiskerprops=dict(color='red'))
plt.show()
```

```python
plt.boxplot(data,capprops=dict(color='green'))
plt.show()
```

```python
plt.boxplot(data,medianprops=dict(color='yellow'))
plt.show()
```

```python
plt.boxplot(data,flierprops=dict(markerfacecolor='red',marker='o'))
plt.show()
```

```python
plt.boxplot(data,whiskerprops=dict(color='red'),boxprops=dict(color='blue'),capprops=dict(color='green'),flierprops=dict(markerfacecolor='red',marker='*'))
plt.show()
```
